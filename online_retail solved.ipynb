## Task 1: Load the Data

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
pip install --upgrade seaborn
```

```python
data = pd.read_excel("Online Retail.xlsx")
data
```

```python
print(data.dtypes)
print(data.shape)
```

## Task 2: Data Cleaning

```python
print(data.isnull().sum())

data_cleaned = data.dropna(subset=['CustomerID'])
print(data_cleaned.shape)
```

## Task 3: Explore the basic statistics of the dataset

```python
# Get basic statistics for numerical columns
print(data_cleaned.describe())

# Display basic statistics for non-numerical columns (e.g., counts of different products)
print(data_cleaned['Description'].value_counts().head())
```

## Task 4: Perform data visualization

```python
# Histogram
plt.figure(figsize=(8,6))
sns.histplot(data_cleaned['Quantity'], bins=30)
plt.title('Distribution of Quantity')
plt.show()

# Bar plot
top_products = data_cleaned['Description'].value_counts().head(10)
plt.figure(figsize=(10,6))
sns.barplot(x=top_products.values, y=top_products.index)
plt.title('Top 10 Products by Sales Quantity')
plt.show()

# Scatter plot
plt.figure(figsize=(8,6))
sns.scatterplot(x='Quantity', y='UnitPrice', data=data_cleaned)
plt.title('Quantity vs Unit Price')
plt.show()
```

## Task 5: Analyze the sales trends over time

```python
# Convert 'InvoiceDate' to datetime
data_cleaned['InvoiceDate'] = pd.to_datetime(data_cleaned['InvoiceDate'])

# Create new columns for month and day of the week
data_cleaned['Month'] = data_cleaned['InvoiceDate'].dt.month
data_cleaned['DayOfWeek'] = data_cleaned['InvoiceDate'].dt.day_name()

# Analyze sales trends by month
monthly_sales = data_cleaned.groupby('Month')['Quantity'].sum()
print("Monthly Sales:")
print(monthly_sales)

# Analyze sales trends by day of the week
daily_sales = data_cleaned.groupby('DayOfWeek')['Quantity'].sum()
print("Sales by Day of the Week:")
print(daily_sales)

# Visualization of sales trends
plt.figure(figsize=(10,6))
sns.barplot(x=monthly_sales.index, y=monthly_sales.values)
plt.title('Sales by Month')
plt.xlabel('Month')
plt.ylabel('Total Quantity Sold')
plt.show()

plt.figure(figsize=(10,6))
sns.barplot(x=daily_sales.index, y=daily_sales.values, order=['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'])
plt.title('Sales by Day of the Week')
plt.xlabel('Day of the Week')
plt.ylabel('Total Quantity Sold')
plt.show()
```

## Task 6: Explore the top-selling products and countries based on the quantity sold

```python
# Top-selling products based on quantity sold
top_products = data_cleaned.groupby('Description')['Quantity'].sum().sort_values(ascending=False).head(10)
print("Top 10 Selling Products:")
print(top_products)

# Top-selling countries based on quantity sold
top_countries = data_cleaned.groupby('Country')['Quantity'].sum().sort_values(ascending=False).head(10)
print("Top 10 Countries by Sales:")
print(top_countries)

# Visualization of top products and countries
plt.figure(figsize=(10,6))
sns.barplot(x=top_products.values, y=top_products.index)
plt.title('Top 10 Selling Products')
plt.xlabel('Total Quantity Sold')
plt.ylabel('Product Description')
plt.show()

plt.figure(figsize=(10,6))
sns.barplot(x=top_countries.values, y=top_countries.index)
plt.title('Top 10 Countries by Sales')
plt.xlabel('Total Quantity Sold')
plt.ylabel('Country')
plt.show()
```

## Task 7: Identify any outliers 

```python
# Visualize distribution of Quantity and UnitPrice to detect outliers
plt.figure(figsize=(10,6))
sns.boxplot(x=data_cleaned['Quantity'])
plt.title('Boxplot of Quantity')
plt.show()

plt.figure(figsize=(10,6))
sns.boxplot(x=data_cleaned['UnitPrice'])
plt.title('Boxplot of Unit Price')
plt.show()

# Find transactions where Quantity or UnitPrice are extreme outliers
outliers_quantity = data_cleaned[(data_cleaned['Quantity'] > data_cleaned['Quantity'].quantile(0.99)) | 
                                 (data_cleaned['Quantity'] < data_cleaned['Quantity'].quantile(0.01))]

outliers_unitprice = data_cleaned[(data_cleaned['UnitPrice'] > data_cleaned['UnitPrice'].quantile(0.99)) | 
                                  (data_cleaned['UnitPrice'] < data_cleaned['UnitPrice'].quantile(0.01))]

print("Outliers in Quantity:")
print(outliers_quantity[['InvoiceNo', 'Description', 'Quantity', 'UnitPrice']].head())

print("Outliers in Unit Price:")
print(outliers_unitprice[['InvoiceNo', 'Description', 'Quantity', 'UnitPrice']].head())
```
